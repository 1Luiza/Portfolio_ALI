```python
a = np.arange(12) # Одномерный массив от 0 до 11
A = a.reshape(4, 3) # Преобразование формы в матрицу 4x3
At = A.T # Транспонирование матрицы A
B = np.dot(A, At)  # Матричное умножение A на At

print("Размер матрицы B:", B.shape)

# Проверка возможности вычисления обратной матрицы для B
try:
    B_inv = np.linalg.inv(B)
    print("Обратная матрица B_inv:\n", B_inv)
except np.linalg.LinAlgError:
    print("Матрица B необратима, поэтому не имеет обратной матрицы.")
```
